```python
arr1 = np.array(one_urls)
arr2 = np.array(two_urls)
arr3 = np.array(three_urls)
arr4 = np.array(four_urls)
arr5 = np.array(five_urls)

# Use concatenate() to join two arrays
all_urls = np.concatenate((arr1, arr2, arr3, arr4, arr5))
len(all_urls)
```

```python
# Retrieve each of the url's HTML data and convert the data into a beautiful soup object. 
# Find, extract and store reviewer names and review text into a list.

names = []
title = []
reviews = []
rating = []
data_string = ""

for url in all_urls:
  params = {'api_key': "9eb445cdfb06c26c57ceb878a755c4ea", 'url': url}
  response = requests.get('http://api.scraperapi.com/',   params=urlencode(params))
  soup = BeautifulSoup(response.text, 'html.parser')

  for item in soup.find_all("span", class_="a-profile-name"):
    data_string = data_string + item.get_text()
    names.append(data_string)
    data_string = "" 
    
  for item in soup.find_all("a", {"data-hook": "review-title"}):
    data_string = data_string + item.getText()
    title.append(data_string)
    data_string = ""
    
  for item in soup.find_all("span", {"data-hook": "review-body"}):
    data_string = data_string + item.get_text()
    reviews.append(data_string)
    data_string = ""
  
  for item in soup.find_all("span", class_="a-icon-alt"):
    data_string = data_string + item.get_text()
    rating.append(data_string[:1])
    data_string = ""
```

```python
# Create the dictionary.
reviews_dict = {'ReviewerName': names, 'ReviewTitle': title, 'Reviews': reviews, 'Rating': rating}
reviews_dict
```

```python
# Delete all the columns that have missing values.
df.dropna(axis=1, inplace=True)
df.head()
```
